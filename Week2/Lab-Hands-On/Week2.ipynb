## Conditional Statements and Decision Structures
Conditional statements allow a program to execute different blocks of code depending on conditions

````python
# -----------------------------------------------------------
# if / elif / else — Age Classification
# -----------------------------------------------------------
# Python evaluates each condition from top to bottom and
# executes ONLY the first matching branch.
# 'elif' (else-if) lets us chain additional conditions
# without nesting. 'else' acts as the default fallback.
# -----------------------------------------------------------
age = 18

if age >= 18:          # Branch 1: adult (18 or older)
    print("You are an adult.")
elif age >= 13:        # Branch 2: teenager (13–17)
    print("You are a teenager.")
else:                  # Branch 3: child (below 13)
    print("You are a child.")
````

````python
# -----------------------------------------------------------
# Simple if — Positive Number Check
# -----------------------------------------------------------
# A standalone 'if' with no 'else' means: do something only
# when the condition is True; otherwise do nothing at all.
# The '>' operator checks strict greater-than (0 excluded).
# -----------------------------------------------------------
a = 3
if a > 0:              # True only when a is strictly positive
    print('a is a positive number')
````

````python
# -----------------------------------------------------------
# if / else — Two-Branch Decision
# -----------------------------------------------------------
# Adding 'else' creates a mandatory fallback that runs
# whenever the 'if' condition is False.
# Here a = 0, so 'a > 0' is False → the else branch runs.
# Note: this else covers BOTH zero and negative values.
# -----------------------------------------------------------
a = 0
if a > 0:
    print('a is a positive number')
else:                  # catches a == 0 and a < 0
    print('a is equals to zero or less than zero')
````

````python
# -----------------------------------------------------------
# if / elif / else — Three Distinct Branches
# -----------------------------------------------------------
# Using 'elif' lets us distinguish zero precisely,
# instead of lumping it together with negatives.
# Python tests conditions in order; only the first True
# branch executes. With a = 0:
#   a > 0  → False  (skip)
#   a < 0  → False  (skip)
#   else   → runs   ✓
# -----------------------------------------------------------
a = 0
if a > 0:
    print('a is a positive number')
elif a < 0:
    print('a is less than zero')
else:                  # only reached when a is exactly 0
    print('a equals to zero')
````

````python
# -----------------------------------------------------------
# Nested Conditional — Positive + Even/Odd Classification
# -----------------------------------------------------------
# The OUTER if/elif/else first classifies the sign of 'a'.
# The INNER if/else (indented deeper) refines the positive
# case by checking whether 'a' is even or odd.
# Key operator: '%' (modulo) returns the division remainder.
#   a % 2 == 0  → no remainder → even number
#   a % 2 != 0  → remainder 1 → odd number
# With a = 4: outer if True, inner if True → 'positive even'
# -----------------------------------------------------------
a = 4
if a > 0:                          # outer: check sign
    if a % 2 == 0:                 # inner: check parity
        print('a is a positive and even integer')
    else:
        print('a is a positive number')
elif a == 0:                       # outer: exact zero check
    print('a equals to zero')
else:                              # outer: negative fallback
    print('a is a negative number')
````

````python
# -----------------------------------------------------------
# Flat Chain with 'and' — Combining Two Conditions
# -----------------------------------------------------------
# Instead of nesting, we use the 'and' logical operator to
# test BOTH the sign and parity in the same condition.
# This keeps the structure flat (no inner if blocks).
# 'and' returns True only when BOTH sub-conditions are True.
# With a = -3:
#   a > 0 and a % 2 == 0  → False (a is not positive)
#   a > 0 and a % 2 != 0  → False
#   a == 0                → False
#   else                  → runs ✓
# -----------------------------------------------------------
a = -3
if a > 0 and a % 2 == 0:           # positive AND even
    print('a is an even and positive integer')
elif a > 0 and a % 2 != 0:         # positive AND odd
    print('a is a positive integer')
elif a == 0:
    print('a equals to zero')
else:                              # negative (any parity)
    print('a is negative')
````

````python
# -----------------------------------------------------------
# Multi-Condition Guard — Access Control
# -----------------------------------------------------------
# Access is only granted when BOTH conditions are satisfied
# simultaneously (using 'and'):
#   1. The username must equal 'admin'
#   2. The access level must be 4 or higher
# Here user = 'Tom' (not 'admin'), so the first condition
# fails immediately and 'Access denied!' is printed.
# -----------------------------------------------------------
user = 'Tom'
access_level = 4

if user == 'admin' and access_level >= 4:  # both must be True
    print('Access granted!')
else:
    print('Access denied!')
````

````python
# -----------------------------------------------------------
# User Input + Role-Based Routing
# -----------------------------------------------------------
# input() reads a string typed by the user.
# .strip() removes any leading/trailing whitespace.
# .lower() normalises to lowercase so 'Admin' and 'ADMIN'
# are both treated the same as 'admin' (case-insensitive).
# -----------------------------------------------------------

# Ask the user to enter their role
role = input("Enter your role (admin/teacher/student): ").strip()

# Check the role and respond
if role.lower() == "admin":
    print("Access granted to admin panel.")
elif role.lower() == "teacher":
    print("Welcome, teacher")
elif role.lower() == "student":
    print("Welcome, student")
else:                              # role string not recognised
    print("Wrong input")
````

````python
# -----------------------------------------------------------
# Real-World Example: Bank Withdrawal Validation
# -----------------------------------------------------------
# Demonstrates chaining multiple elif guards to protect
# against invalid inputs before processing a transaction.
# float() converts the input string to a decimal number
# so we can compare monetary values correctly.
# The '-=' operator subtracts the withdrawal in place:
#   balance -= withdrawal  is shorthand for
#   balance = balance - withdrawal
# The f-string (f"...") allows embedding variables directly
# inside a string using curly braces {balance}.
# -----------------------------------------------------------

# Ask the user for their current balance
balance = float(input("Enter your current account balance: "))

# Ask the user how much they want to withdraw
withdrawal = float(input("Enter the amount you want to withdraw: "))

# Guard 1: withdrawal must be a positive amount
if withdrawal <= 0:
    print("Withdrawal amount must be greater than 0.")
# Guard 2: cannot withdraw more than the current balance
elif withdrawal > balance:
    print("Insufficient funds. Transaction denied.")
else:
    balance -= withdrawal          # deduct the withdrawal
    print(f"Transaction approved. Your new balance is: {balance:}")
````

````python
# -----------------------------------------------------------
# Real-World Example: School Enrollment Eligibility
# -----------------------------------------------------------
# Illustrates NESTED conditionals in a practical scenario.
# The outer if/elif/else classifies by age group.
# The inner if/else (inside the school-age branch) adds a
# secondary validation (credits cannot be negative).
# int() converts the input string to a whole number.
# -----------------------------------------------------------

# Ask the user for their age
age = int(input("Enter your age: "))

# Ask the user for the number of credits they have completed
credits = int(input("Enter the number of credits you have completed: "))

# Outer level: classify by age group
if age < 5:                        # too young for school
    print("You are too young to enroll.")
elif age >= 5 and age <= 18:       # standard school age
    # Inner level: validate the credits value
    if credits >= 0:
        print("You are eligible for school enrollment.")
    else:                          # negative credits are invalid
        print("Credits cannot be negative.")
else:                              # adult (over 18)
    print("You may be eligible for adult education or higher studies.")
````

## Shorthand Conditional (Ternary Operator) in Python

**Code Example**
```python
number = 3

# Shorthand if-else (ternary operator)
print('Next number should be', number + 1) if number != 5 else print("loop's end")
```

**Explanation**
- expression1 → executed if the condition is True
- expression2 → executed if the condition is False

number != 5 → condition being tested

print('Next number should be', number + 1) → executed if number is not 5

print("loop's end") → executed if number is 5

````python
# -----------------------------------------------------------
# Ternary (Inline) Conditional
# -----------------------------------------------------------
# Python's ternary form collapses a two-branch if/else into
# a single expression:
#   <value_if_true> if <condition> else <value_if_false>
# Here the condition is  number != 5.
# Since number = 3, the condition is True, so the left side
# (print 'Next number should be 4') executes.
# Ternary expressions are best for simple, readable cases.
# -----------------------------------------------------------
number = 3
print('Next number should be', number + 1) if number != 5 else print("loop's end")
````

## Loops in Programming
A **loop** is a control flow structure that repeatedly executes a block of code as long as a specified condition is true or for a predefined number of iterations. Loops are fundamental for automating repetitive tasks, reducing code duplication, and managing iterative computations.


### **For Loop**
   
**Example:**
```python
# For loop: iterates over a sequence (list, tuple, string, range)
for i in range(5):
    print(i)  # prints 0 to 4
```
**Usage:**
- Iterating through elements of a list
- Generating a fixed number of repetitions
- Creating new lists using list comprehensions

````python
# -----------------------------------------------------------
# for loop — Iterating Over a List
# -----------------------------------------------------------
# A 'for' loop visits each element of a sequence in order.
# 'number' is a loop variable — a temporary name that takes
# the value of each item one at a time. It only exists
# inside the loop body (the indented block below 'for').
# -----------------------------------------------------------
numbers = [0, 1, 2, 3, 4, 5]
for number in numbers:   # number is temporary name to refer to the list's items, valid only inside this loop 
    print(number)
````

````python
# -----------------------------------------------------------
# for loop — Iterating Over a String (Character by Character)
# -----------------------------------------------------------
# Strings in Python are iterable sequences of characters.
# A 'for' loop can traverse a string the same way it
# traverses a list — one element (character) at a time.
# -----------------------------------------------------------
language = 'Python'
for letter in language:        # 'letter' holds each character
    print(letter)
````

````python
# -----------------------------------------------------------
# for loop — Index-Based String Traversal
# -----------------------------------------------------------
# An alternative approach: iterate over a range of indices
# and use each index 'i' to access characters by position.
#   len(language) → returns the length of the string (10)
#   range(10)     → generates [0, 1, 2, ..., 9]
#   language[i]   → accesses the character at position i
# This approach is useful when you also need the index,
# e.g. to compare adjacent characters or print positions.
# -----------------------------------------------------------
language = 'Javascript'
for i in range(len(language)):   # i takes values 0, 1, ..., 9
    print(language[i])           # index into the string
````

````python
# -----------------------------------------------------------
# for loop — Iterating Over a Dictionary
# -----------------------------------------------------------
# Dictionaries store data as key-value pairs.
# Iterating over a dict directly yields its KEYS only.
# Using .items() yields (key, value) tuples, letting you
# unpack both in the loop variable list: 'for key, value'.
# The nested dict 'person_info' and list 'children' show
# that dictionary values can be any Python object.
# -----------------------------------------------------------
actor = {
    'first_name': 'Tom',
    'last_name': 'Hanks',
    'country': 'United States',
    'city': 'Concord, California',
    'age': 69,
    'is_married': False,
    'skills': ['Comedy Actor', 'Drama Actor', 'Filmmaker'],
    'person_info': {
        'firstname': 'Tom',
        'lastname': 'Hanks',
        'country': 'United States',
        'city': 'Concord, California',
        'children': [' Colin Hanks', 'Elizabeth Hanks', 'Chet Hanks', 'Truman Hanks']
    }
}

# Loop 1: yields only KEYS
for key in actor:
    print(key)

print("====================================================================================================")

# Loop 2: .items() yields (key, value) pairs
for key, value in actor.items():
    print(key, value)
````

### **Nested Loops**
**Usage:**
- Iterating over multi-dimensional data (e.g., matrices, grids)
- Combining multiple sequences

````python
# -----------------------------------------------------------
# Nested for loops — Grid / Coordinate Pairs
# -----------------------------------------------------------
# For each value of the OUTER loop variable 'i' (0, 1, 2),
# the INNER loop runs its FULL cycle through 'j' (0, 1).
# This produces every combination of (i, j) — essentially
# generating a 3×2 grid of coordinate pairs.
# Total iterations = len(range(3)) × len(range(2)) = 6
# -----------------------------------------------------------
for i in range(3):       # outer loop: 0, 1, 2
    for j in range(2):   # inner loop: 0, 1  (runs fully for each i)
        print(i, j)
````

### **Loop Control Statements: break and continue**

#### **break**
- **Purpose:** Immediately exits the loop, even if the loop condition has not finished.
- **Use case:** Stop iterating when a specific condition is met.

````python
# -----------------------------------------------------------
# break — Early Exit from a for Loop
# -----------------------------------------------------------
# 'break' immediately terminates the loop, jumping past
# the remaining iterations and any 'else' clause.
# Here we iterate over a tuple (immutable ordered sequence).
# When number == 3, we print it, then 'break' fires —
# so 4 and 5 are never reached.
# Output: 0, 1, 2, 3
# -----------------------------------------------------------
numbers = (0, 1, 2, 3, 4, 5)   # tuple — same syntax as list but immutable
for number in numbers:
    print(number)
    if number == 3:
        break                  # stop the loop immediately
````

#### **continue**
- **Purpose:** Skips the current iteration and moves to the next iteration of the loop.
- **Use case:** Skip processing certain elements while continuing the loop.

````python
# -----------------------------------------------------------
# continue — Skip Even Numbers, Print Only Odds
# -----------------------------------------------------------
# 'continue' skips the REST of the current loop body and
# jumps straight to the next iteration — the loop does NOT
# terminate (unlike 'break').
# When i % 2 == 0 (even), we skip the print statement.
# When i % 2 != 0 (odd), the if is False, so we reach print.
# range(1, 10) generates 1 through 9 (10 is excluded).
# -----------------------------------------------------------
for i in range(1, 10):
    if i % 2 == 0:
        continue           # skip even numbers — jump to next i
    print("Odd number:", i)
````

````python
# -----------------------------------------------------------
# list() and set() from range()
# -----------------------------------------------------------
# range() by itself is a lazy sequence — it does not store
# all values in memory at once. Wrapping it in list() or
# set() forces it to generate all values immediately.
#
# list(range(11))   → [0, 1, 2, ..., 10]  (ordered, includes 0)
# set(range(1, 11)) → {1, 2, ..., 10}    (unordered, no duplicates)
#
# Note: sets do not guarantee display order — the printed
# output may appear in a different arrangement each run.
# -----------------------------------------------------------
numbers = list(range(11))           # list: ordered, 0 to 10
print(numbers)
numbers_set = set(range(1, 11))     # set: unique values, 1 to 10
print(numbers_set)
````

````python
# -----------------------------------------------------------
# for / else — The 'else' Clause on a Loop
# -----------------------------------------------------------
# Python's loops support an optional 'else' block that runs
# ONLY when the loop finishes naturally (i.e. was NOT exited
# early via 'break'). This is useful to detect whether a
# search or process completed without interruption.
# Here there is no 'break', so the else block always runs.
# After the loop, 'number' retains its last value (10).
# -----------------------------------------------------------
for number in range(11):
    print(number)               # prints 0 to 10 (11 is excluded)
else:
    # Runs because the loop completed without a 'break'
    print('The loop stops at', number)
````

````python
# -----------------------------------------------------------
# Nested Loop — Drilling Into a Dictionary's List Value
# -----------------------------------------------------------
# The outer loop iterates over the dictionary's keys.
# When we reach the 'skills' key, the inner loop iterates
# over the LIST stored as its value, printing each skill.
# This pattern is common when a dict contains mixed types
# (strings, numbers, lists, nested dicts) as values.
# -----------------------------------------------------------
actor = {
    'first_name': 'Tom',
    'last_name': 'Hanks',
    'country': 'United States',
    'city': 'Concord, California',
    'age': 69,
    'is_married': False,
    'skills': ['Comedy Actor', 'Drama Actor', 'Filmmaker'],
    'person_info': {
        'firstname': 'Tom',
        'lastname': 'Hanks',
        'country': 'United States',
        'city': 'Concord, California',
        'children': [' Colin Hanks', 'Elizabeth Hanks', 'Chet Hanks', 'Truman Hanks']
    }
}

for key in actor:              # outer: iterate over dict keys
    if key == 'skills':
        for skill in actor['skills']:   # inner: iterate over the skills list
            print(skill)
````

````python
# -----------------------------------------------------------
# Generator Expression with Conditional Filter
# -----------------------------------------------------------
# sum() can accept a generator expression — a compact way
# to apply a filter and aggregation in one line.
# Syntax: (expression for item in iterable if condition)
#
# Here: iterate -10 to 9, keep only positive numbers (>0),
# and sum them. Positives: 1+2+3+4+5+6+7+8+9 = 45.
# The docstring below shows the step-by-step trace.
# -----------------------------------------------------------
total = sum(number for number in range(-10, 10) if number > 0)
print("Total:", total)

# Tracing showing which values are skipped vs added:
"""Skipping -10, not positive
Skipping -9, not positive
Skipping -8, not positive
Skipping -7, not positive
Skipping -6, not positive
Skipping -5, not positive
Skipping -4, not positive
Skipping -3, not positive
Skipping -2, not positive
Skipping -1, not positive
Skipping 0, not positive
Adding 1, running total: 1
Adding 2, running total: 3
Adding 3, running total: 6
Adding 4, running total: 10
Adding 5, running total: 15
Adding 6, running total: 21
Adding 7, running total: 28
Adding 8, running total: 36
Adding 9, running total: 45

Final total: 45"""
````

### **While Loop**
Executes as long as a condition is true.

**Example:**
```python
count = 0
while count < 5:
    print(count)
    count += 1
```
>**Usage:** Automating repetitive tasks, such as processing each item in a list.

````python
# -----------------------------------------------------------
# while / else — Count to 5 with Else Clause
# -----------------------------------------------------------
# A 'while' loop repeats its body as long as the condition
# (count <= 5) remains True. We MUST update 'count' inside
# the loop; forgetting to do so creates an infinite loop.
#
# The 'else' block runs after the while condition becomes
# False (i.e. the loop ended naturally, not via 'break').
# At that point count = 6, which is printed by else.
# Output: 0 1 2 3 4 5 6
# -----------------------------------------------------------
count = 0
while count <= 5:          # loop continues while count is 0–5
    print(count)
    count = count + 1      # increment; prevents infinite loop
else:
    print(count)           # count is now 6 (condition just failed)
````

````python
# -----------------------------------------------------------
# while + break — Early Exit from a While Loop
# -----------------------------------------------------------
# The loop condition allows count up to 5, but 'break'
# terminates the loop early when count reaches 3.
# Sequence of events each iteration:
#   1. Check while condition (count <= 5)
#   2. Print count
#   3. Increment count
#   4. If count == 3 → break (loop exits immediately)
# Output: 0 1 2  (3 is never printed — see trace table below)
# -----------------------------------------------------------
count = 0
while count <= 5:
    print(count)
    count = count + 1
    if count == 3:
        break              # exit the loop; else block is skipped
````

### Execution Trace Table of the above code

| Iteration | count at Start | Condition (count <= 5) | Printed Value | count After Increment | count == 3 | Break Executed | Loop Continues? |
|-----------|----------------|------------------------|---------------|-----------------------|------------|----------------|-----------------|
| 1         | 0              | True                   | 0             | 1                     | False      | No             | Yes             |
| 2         | 1              | True                   | 1             | 2                     | False      | No             | Yes             |
| 3         | 2              | True                   | 2             | 3                     | True       | Yes            | No              |

````python
# -----------------------------------------------------------
# while + continue — Skip a Specific Value
# -----------------------------------------------------------
# When count == 3, we increment FIRST (to avoid getting
# stuck on 3 forever) and then 'continue' skips the print.
# Unlike 'break', 'continue' only skips the REST of the
# current body — the loop keeps running.
# Output: 0 1 2 4  (3 is skipped)
#
# Important: always increment before 'continue' in a
# while loop, or the condition never changes → infinite loop.
# -----------------------------------------------------------
count = 0
while count < 5:
    if count == 3:
        count += 1         # must increment BEFORE continue
        continue           # skip the print below, go back to while
    print(count)
    count = count + 1
````

##  Basic List Comprehensions as Compact Logic Expressions

### Definition

A **list comprehension** is a concise way to create a list by combining:

- iteration
- optional conditional filtering
- an output expression

It replaces longer `for` loop constructions with a single, readable line.

### Key Characteristics

- More concise than traditional loops.
- Improves readability for simple transformations.
- Combines iteration and logic in a single expression.
- Suitable for simple transformations and filtering.
- For complex multi-step logic, a standard loop is preferable.


### General Syntax

```python
[expression for item in iterable if condition]
```

### Components

- `expression` → the value added to the new list  
- `item` → the variable representing each element  
- `iterable` → the sequence being iterated over  
- `if condition` → optional filter to include only certain elements  

### Computational Insight

A list comprehension:

```
[expression for item in iterable if condition]
```

is conceptually equivalent to:

```
for item in iterable:
    if condition:
        result.append(expression)
```

but executed in a more compact and often more efficient manner.

````python
# -----------------------------------------------------------
# List Comprehension — Squaring Numbers
# -----------------------------------------------------------
# [x * x for x in range(5)] builds a new list by computing
# x squared for every x in {0, 1, 2, 3, 4}.
# The standard loop equivalent is shown in the docstring.
# Note: x**2 and x*x are equivalent; both express squaring.
# -----------------------------------------------------------
squares = [x * x for x in range(5)]   # expression: x*x, iterable: range(5)
print(squares)

"""
Standard Loop equivalent:
squares = []
for x in range(5):
    squares.append(x * x)
"""
````

````python
# -----------------------------------------------------------
# List Comprehension — Filtering with a Condition
# -----------------------------------------------------------
# The 'if x % 2 == 0' part acts as a FILTER — only even
# numbers (those with no remainder when divided by 2) are
# included in the output list. Odd numbers are discarded.
# This is different from the ternary ('if/else') form used
# in the next cell: a trailing 'if' FILTERS; a mid-expression
# 'if/else' TRANSFORMS (keeps all items, changes their value).
# -----------------------------------------------------------
evens = [x for x in range(10) if x % 2 == 0]   # filter: keep even only
print(evens)

"""
Standard Loop equivalent:
evens = []
for x in range(10):
    if x % 2 == 0:
        evens.append(x)
"""
````

````python
# -----------------------------------------------------------
# List Comprehension — Inline if/else (Ternary Transform)
# -----------------------------------------------------------
# Placing 'if/else' BEFORE 'for' is a TRANSFORM, not a filter:
# every item is kept, but its value is replaced depending
# on the condition.
# Syntax: [<true_val> if <cond> else <false_val> for x in ...]
#
# For each x in 0–4:
#   x % 2 == 0  → label is "even"
#   x % 2 != 0  → label is "odd"
# All 5 numbers are represented; only their label differs.
# -----------------------------------------------------------
labels = ["even" if x % 2 == 0 else "odd" for x in range(5)]
print(labels)

"""
Standard Loop equivalent:
labels = []
for x in range(5):
    if x % 2 == 0:
        labels.append("even")
    else:
        labels.append("odd")
"""
````
