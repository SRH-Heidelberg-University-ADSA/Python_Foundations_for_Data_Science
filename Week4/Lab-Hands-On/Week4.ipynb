# Functions & Modular Programming
This notebook introduces Python functions and modular programming, covering parameters, return values, variable scope, default arguments, and lambda expressions. Each section includes explanations and examples.

## **1. Purpose of Functions**
Functions are reusable blocks of code that perform a specific task. They help:  
- Avoid *code repetition*  
- Make code *modular and readable*
- Allow *easy debugging and testing*

**Example:**

```python
def greet(name):
    """Print a greeting message for the given name."""
    print(f"Hello, {name}!")

greet("Alice")
greet("Bob")
```

## **2. Parameters and Return Values**
- **Parameters** are inputs to a function.
-  **Return values** allow a function to produce output that can be stored or used later.
  
**Example:**

```python
def square(x):
    return x ** 2

result = square(5)
print(result)  # Output: 25
``` 

````python
# define a function to generate and print a full name
def generate_full_name():
    # assign the first name
    first_name = 'Tom'
    
    # assign the last name
    last_name = 'Hanks'
    
    # define a space character to separate names
    space = ' '
    
    # concatenate first name, space, and last name to form the full name
    full_name = first_name + space + last_name
    
    # print the full name
    print(full_name)

# call the function to execute it
generate_full_name()
````

````python
# define a function that generates and returns a full name
def return_generate_full_name():
    # assign the first name
    first_name = 'Tom'
    
    # assign the last name
    last_name = 'Cruise'
    
    # define a space character to separate names
    space = ' '
    
    # concatenate first name, space, and last name to form the full name
    full_name = first_name + space + last_name
    
    # return the full name instead of printing it
    return full_name

# call the function and print the returned full name
print(return_generate_full_name())
````

````python
# define a function that generates a greeting message for a given name
def greetings(name):
    # create a personalised message by concatenating the name with a welcome string
    message = name + ', welcome to SRH!'
    
    # return the constructed message
    return message

# prompt the user to enter their name
name = input("please enter your name:")

# call the function and print the returned greeting message
print(greetings(name))
````

````python
# define a function that adds ten to a given number
def add_ten(num):
    # assign the constant value ten
    ten = 10
    
    # return the result of adding ten to the input number
    return num + ten

# call the function with the value 90 and print the result
print(add_ten(90))
````

````python
# define a function that checks whether a number is even
def is_even(n):
    # check if the remainder when dividing by 2 is zero
    if n % 2 == 0:
        # return True if the number is even
        # note: return exits the function immediately once executed
        return True
    
    # return False if the number is not even (i.e., odd)
    return False

# call the function with 100 and print the result (expected: True)
print(is_even(100)) 

# call the function with 11 and print the result (expected: False)
print(is_even(11))
````

````python
# define a function that returns all even numbers from 0 up to n (inclusive)
def find_even_numbers(n):
    # initialise an empty list to store even numbers
    evens = []
    
    # iterate through all integers from 0 to n inclusive
    for i in range(n + 1):
        # check if the current number is even
        if i % 2 == 0:
            # append the even number to the list
            evens.append(i)
    
    # return the list of even numbers
    return evens

# call the function with 22 and print the resulting list
print(find_even_numbers(22))
````

## **3. Variable Scope**
- **Local variables** exist inside a function.
-  **Global variables** exist outside and can be accessed anywhere, but modifying them inside a function requires **global** keyword.
  
**Example:**

```python
x = 10  # global

def modify():
    y = 5  # local
    print("Local y:", y)

modify()
print("Global x:", x)
``` 

## **4. Default Parameters**
Functions can have **default values** for parameters if no argument is provided.

**Example:**

```python
def greet(name="Guest"):
    print(f"Hello, {name}!")

greet("Alice")  # Hello, Alice!
greet()         # Hello, Guest!
``` 

````python
# define a function to calculate age based on birth year and current year
# default values are provided if no arguments are passed
def calculate_age(birth_year=2000, current_year=2026):
    # compute the age by subtracting birth year from current year
    age = current_year - birth_year
    
    # return the calculated age
    return age 

# call the function without arguments to use default values
print('Default Age:', calculate_age())  

# prompt the user to enter their year of birth
dob = input("please enter your date of birth:")

# prompt the user to enter the current year
current_year = input("please enter the current year:")

# convert inputs to integers and call the function with user-provided values
print('Age:', calculate_age(int(dob), int(current_year)))
````

````python
# arbitrary number of arguments
'''
if the number of arguments is not known in advance, a function can accept
a variable number of positional arguments by using * before the parameter name
'''

# define a function that sums an arbitrary number of numeric inputs
def sum_all_nums(*nums):
    # initialise a variable to store the cumulative sum
    total = 0
    
    # iterate over each value passed to the function
    for num in nums:
        # add the current number to the total
        total += num    
    
    # return the final computed sum
    return total

# call the function with multiple arguments and print the result
print(sum_all_nums(2, 3, 5, 6))

# call the function with a different number of arguments
print(sum_all_nums(10, 12))
````

````python
# define a function that accepts a film name and a variable number of actors
def generate_actors(film, *args):
    # print the movie name
    print("Movie name:", film)
    
    # print a header for the cast list
    print("The cast:")
    
    # iterate over all actors passed as arbitrary arguments
    for i in args:
        # print each actor's name
        print(i)

# call the function with a film title and multiple actor names
generate_actors(
    'The Terminal',
    'Tom Hanks',
    'Diego Luna',
    'Zoe Saldaña',
    'Catherine Zeta-Jones'
)
````

````python
# dictionary unpacking
# define a dictionary containing key-value pairs that match the function parameters
harry_in_london = {
    "name": "Harry",
    "location": "London",
    "performance": "2010"
}

# define a function that accepts named parameters
def greet(name, location, performance):
    # print a formatted introduction using f-string
    print(f"Hi, I am {name}. I am from {location}. I have been performing since {performance}.")

# call the function using dictionary unpacking
# the ** operator unpacks the dictionary into keyword arguments
# each key in the dictionary must match a parameter name in the function
greet(**harry_in_london)
````

````python
# define a function that accepts an arbitrary number of keyword arguments
def onthefly_dictionary(**args):
    # print the number of key-value pairs passed to the function
    print("The length of the on the fly dictionary is:", len(args))
    
    # verify that the received object is of type dictionary
    print("It is a dictionary, see:", type(args))
    
    # print a header for displaying the dictionary contents
    print("Let's print the on the fly dictionary:")
    
    # iterate over key-value pairs in the dictionary
    for k, v in args.items():
        # print each key and its corresponding value
        print(" * key:", k, "==>", "value:", v)
    
    # return the constructed dictionary
    return args


# call the function with keyword arguments to create a dictionary dynamically
person = onthefly_dictionary(
    first_name='Tom',
    last_name='Hanks',
    age=20,
    country='USA'
)

# print a separator line for clarity
print("\nSaved dictionary:")

# print the returned dictionary stored in the variable 'person'
print(person)
````

## **5. Lambda Expressions**
- Anonymous functions for short operations.
- Typically one-liner with `lambda arguments: expression`

**Example:**

```python
square = lambda x: x ** 2
print(square(6))  # Output: 36

add = lambda a, b: a + b
print(add(3, 5))  # Output: 8
``` 

````python
# option 1: define a standard function that returns the sum of two numbers
def add_two_nums(a, b):
    # return the result of adding a and b
    return a + b

# call the function and print the result
print(add_two_nums(2, 3))  


# option 2: define the same functionality using a lambda (anonymous function)
# lambda functions are concise, single-expression functions without a formal name
add_two_nums = lambda a, b: a + b

# call the lambda function and print the result
print(add_two_nums(2, 3))   


# self-invoking lambda function
# define and immediately execute the lambda with arguments (2, 3)
print((lambda a, b: a + b)(2, 3))


# define a lambda function to compute the square of a number
square = lambda x: x ** 2

# call the lambda function and print the result
print(square(3))   


# define a lambda function to compute the cube of a number
cube = lambda x: x ** 3

# call the lambda function and print the result
print(cube(3))
````

````python
# explanation:
'''
- first, power(2) returns a lambda function that raises 2 to any power
- then, this returned lambda function is immediately called with the argument 3, computing 2 ** 3
- the result (8) is assigned to the variable cube
'''

# define a function that returns another function (closure)
def power(x):
    # return a lambda function that takes n and computes x raised to the power n
    return lambda n: x ** n

# call power(2) to create a function that raises 2 to a given power,
# then immediately call it with 3
cube = power(2)(3)

# print the result (expected: 8)
print(cube)
````
