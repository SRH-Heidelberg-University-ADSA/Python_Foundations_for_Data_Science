# Variables and Basic Data Types
Understanding data types is fundamental for data preprocessing and modelling tasks.

### Core Data Types:
| Type | Description | Example |
|---|---|---|
| `int` | Integer values | `69` |
| `float` | Real (decimal) numbers | `3.14` |
| `str` | Text | `'Hello'` |
| `bool` | True or False | `True` |
| `list` | Ordered, mutable collection | `[1, 2, 3]` |
| `tuple` | Ordered, immutable collection | `(1, 2, 3)` |
| `dict` | Key-value mapping | `{'name': 'Tom'}` |

Use `type()` to inspect any variable's type.

```python
# Declaring variables of different types
first_name  = 'Tom'                   # str
last_name   = 'Hanks'                 # str
country     = 'United States'         # str
city        = 'Concord, California'   # str
age         = 69                      # int
height      = 1.83                    # float
is_married  = False                   # bool
skills      = ['Comedy Actor', 'Drama Actor', 'Filmmaker']            # list
films       = ('The Terminal', 'Here', 'A Man Called Otto')           # tuple
person_info = {'firstname': 'Tom', 'lastname': 'Hanks',
               'country': 'United States', 'city': 'Concord, California'}  # dict
```

```python
# Printing variables and their types using type()
print("First name:", first_name,   "→ type:", type(first_name))
print("Last name: ", last_name,    "→ type:", type(last_name))
print("Age:       ", age,          "→ type:", type(age))
print("Height:    ", height,       "→ type:", type(height))
print("Married:   ", is_married,   "→ type:", type(is_married))
print("Skills:    ", skills,       "→ type:", type(skills))
print("Films:     ", films,        "→ type:", type(films))
print("Info dict: ", person_info,  "→ type:", type(person_info))
```

### Modern String Formatting: f-strings
f-strings (formatted string literals) are the **recommended modern way** to embed variables directly in strings. They are cleaner and faster than `.format()` or `%` formatting.

Syntax: prefix the string with `f` and wrap variables in `{}`.

```python
# TODO-Add your firstname
first_name=""
# TODO-Add your lastname
last_name= ""
# TODO-Add your age
age=0
# TODO-Add your country
country= ""
# TODO-Add your height
height=0
# TODO-Add your is_married
is_married=False

# f-string examples
print(f"Hello, my name is {first_name} {last_name}.")       # embed variables
print(f"I am {age} years old and from {country}.")          # multiple variables
print(f"My height is {height:.1f} metres.")                 # format float to 1 decimal
print(f"Married: {is_married}")                             # bool works too

# Comparison: old .format() style vs f-string
old_style = "I am {} {}.".format(first_name, last_name)
new_style = f"I am {first_name} {last_name}."
print("\nOld .format() style:", old_style)
print("New f-string style: ", new_style)
```

```python
# Declaring multiple variables in one line (tuple unpacking)
x, y, z = 10, 20, 30
print(f"x={x}, y={y}, z={z}")

# Swapping variables (Python makes this elegant!)
x, y = y, x
print(f"After swap: x={x}, y={y}")
```

### Essential String Operations
Strings are sequences of characters. Here are the most important operations for a Week 1 foundation.

```python
language = 'Python'

# Length
print(f"Length of '{language}': {len(language)}")

# Indexing (0-based; negative counts from the end)
print(f"First character  [0]:  {language[0]}")    # P
print(f"Last character   [-1]: {language[-1]}")   # n

# Slicing [start:stop:step]
print(f"First 3 chars  [0:3]: {language[0:3]}")   # Pyt
print(f"Last 3 chars   [-3:]: {language[-3:]}")   # hon
print(f"Every 2nd char [::2]: {language[::2]}")   # Pto

# Concatenation
full_name = first_name + ' ' + last_name
print(f"Concatenated: {full_name}")
```

```python
# Key string methods useful from Day 1
sentence = '  hello, world!  '

print(sentence.strip())           # remove leading/trailing whitespace → 'hello, world!'
print(sentence.strip().upper())   # uppercase → 'HELLO, WORLD!'
print(sentence.strip().title())   # title case → 'Hello, World!'
print(sentence.strip().replace('world', 'Python'))  # replace → 'hello, Python!'
print('hello world'.split())      # split on whitespace → ['hello', 'world']
# The string ', ' acts as a separator. The method .join(...) is a string method that concatenates elements of an iterable (here, a list)
print(', '.join(['Tom', 'Ana', 'Leo']))  # join list into string → Tom, Ana, Leo
print(' '.join(['Tom', 'Ana', 'Leo']))  # join list into string → Tom Ana Leo
```

# Arithmetic Operations
Python supports all standard arithmetic operators.

| Operator | Name | Example | Result |
|---|---|---|---|
| `+` | Addition | `7 + 2` | `9` |
| `-` | Subtraction | `7 - 2` | `5` |
| `*` | Multiplication | `7 * 2` | `14` |
| `/` | True division | `7 / 2` | `3.5` |
| `//` | Floor division | `7 // 2` | `3` |
| `%` | Modulus (remainder) | `7 % 2` | `1` |
| `**` | Exponentiation | `7 ** 2` | `49` |

```python
a, b = 7, 2

print(f"{a} + {b}  = {a + b}   (Addition)")
print(f"{a} - {b}  = {a - b}   (Subtraction)")
print(f"{a} * {b}  = {a * b}  (Multiplication)")
print(f"{a} / {b}  = {a / b}  (True Division — always returns float)")
print(f"{a} // {b} = {a // b}   (Floor Division — drops decimal)")
print(f"{a} % {b}  = {a % b}   (Modulus — remainder only)")
print(f"{a} ** {b} = {a ** b}  (Exponentiation)")
```

```python
# Understanding Modulo visually
# 7 % 3 → "How much is LEFT after dividing 7 by 3 as many times as possible?"
# 7 = 3 * 2 + 1  →  remainder is 1

dividend  = 7
divisor   = 3
quotient  = dividend // divisor   # how many full times divisor fits
remainder = dividend %  divisor   # what's left over

print(f"{dividend} = {divisor} × {quotient} + {remainder}")
print(f"So {dividend} % {divisor} = {remainder}")

# Real-world use: checking even/odd
for n in [4, 7, 12, 15]:
    parity = 'even' if n % 2 == 0 else 'odd'
    print(f"{n} is {parity}")
```

```python
# Real-world arithmetic examples

# Area of a circle
radius = 10
pi = 3.14159
area = pi * radius ** 2
print(f"Area of circle (r={radius}): {area:.2f}")

# Weight of an object
mass    = 75     # kg
gravity = 9.81   # m/s²
weight  = mass * gravity
print(f"Weight of {mass}kg object: {weight:.2f} N")

# Operator precedence reminder: PEMDAS
result = (5 + 3) * 2 ** 3 - 4 / 2
print(f"(5+3) * 2**3 - 4/2 = {result}")
```

# Boolean Comparisons & Logical Operators
A boolean comparison always evaluates to `True` or `False`.

### Comparison Operators
| Operator | Meaning | Example | Result |
|---|---|---|---|
| `==` | Equal to | `3 == 3` | `True` |
| `!=` | Not equal | `3 != 2` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<` | Less than | `2 < 1` | `False` |
| `>=` | Greater than or equal | `5 >= 5` | `True` |
| `<=` | Less than or equal | `4 <= 3` | `False` |

### Logical Operators
| Operator | Meaning | Rule |
|---|---|---|
| `and` | Both must be True | `True and False → False` |
| `or` | At least one must be True | `True or False → True` |
| `not` | Flip the value | `not True → False` |

> **Common mistake:** `=` is assignment. `==` is comparison. Mixing them up is one of the most common beginner bugs!

```python
# Comparison operators
x, y = 10, 15

print(f"x={x}, y={y}")
print(f"x > y  → {x > y}")
print(f"x < y  → {x < y}")
print(f"x == 10 → {x == 10}")
print(f"y != 15 → {y != 15}")
print(f"x >= 5  → {x >= 5}")
print(f"y <= 20 → {y <= 20}")
```

```python
# Logical operators
a, b = 8, 12

print(f"a={a}, b={b}")
print(f"a < 10 and b > 10 → {a < 10 and b > 10}")
print(f"a > 5  or  b < 10 → {a > 5  or  b < 10}")
print(f"not (a == 8)       → {not (a == 8)}")

# Chained comparisons — Pythonic and readable
num = 7
print(f"\n5 < {num} < 10 → {5 < num < 10}")
```

```python
# empty string
print(f"bool("")",bool(""))
# one character - Whitespace
print(f"bool(" ")",bool(" "))
```

# Data Type Conversions (Type Casting)
Data science workflows frequently require explicit type casting during data cleaning and preprocessing.

### Why it matters:
Data read from files (CSV, JSON, APIs) often arrives as **strings**, even when the values are numbers. You must cast them before doing any computation.

### Type Casting Functions
| Function | Converts to | Example |
|---|---|---|
| `int()` | Integer | `int("20")` → `20` |
| `float()` | Float | `float("3.5")` → `3.5` |
| `str()` | String | `str(42)` → `"42"` |
| `bool()` | Boolean | `bool(0)` → `False` |

```python
# Why casting matters — this is a common real-world bug!
user_input = "20"          # simulates reading from a file or user input

# This will crash:
# result = user_input + 5  # TypeError: can only concatenate str to str

# Fix: cast first
user_input_int = int(user_input)
result = user_input_int + 5
print(f"Casting '20' (str) to int: {user_input_int}, type: {type(user_input_int)}")
print(f"After adding 5: {result}")
```

```python
# int() truncates (does NOT round) — important nuance!
print(f"int(44.4) = {int(44.4)}")   # 44  (drops decimal)
print(f"int(44.9) = {int(44.9)}")   # 44  (still 44, not 45!)
print(f"int(44.9) is not the same as round(44.9) = {round(44.9)}")

# float() → int and back
s1 = "20"
s2 = "3.5"
n1 = int(s1)
n2 = float(s2)
print(f"\n'{s1}' → int: {n1},  type: {type(n1)}")
print(f"'{s2}' → float: {n2}, type: {type(n2)}")
```

```python
# str() — number to string for concatenation
num = 42
# message = "The answer is " + num  # TypeError!
message = "The answer is " + str(num)
print(message)

# Or better: just use an f-string!
print(f"The answer is {num}")
```

```python
# bool() — truthiness rules
# Falsy: 0, 0.0, "", [], {}, None
# Truthy: everything else

print("--- Integers ---")
print(f"bool(0)  = {bool(0)}")
print(f"bool(1)  = {bool(1)}")
print(f"bool(-5) = {bool(-5)}")

print("\n--- Strings ---")
print(f"bool('')       = {bool('')}")
print(f"bool('Python') = {bool('Python')}")

print("\n--- Collections ---")
print(f"bool([])  = {bool([])}")
print(f"bool([1]) = {bool([1])}")
```

```python
# Variable Reassignment — integers are immutable, but variables are just labels!
x = 5
print(f"x = {x}, id(x) = {id(x)}")  # id() shows the memory address of the object

x = 10
print(f"x = {x}, id(x) = {id(x)}")  # different address — x now points to a new object

# The integer 5 itself didn't change — x just re-labelled to point at 10.
# This is why Python says integers are immutable:
# the VALUE 5 cannot be modified, but the VARIABLE x can be redirected.
```
