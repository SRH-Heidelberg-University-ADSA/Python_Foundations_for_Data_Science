# 1. Lists
**Definition:**  
Lists are ordered, mutable sequences that allow storing multiple items in a single variable.

```python
numbers = [10, 20, 30, 40]
```

**Key Features**:

- Ordered (elements maintain insertion order)

- Mutable (elements can be added, removed, or modified)

- Indexed (zero-based indexing, supports negative indices and slicing)


````python
# empty list
empty_list = list()  
print(len(empty_list))  
````

````python
fruits = ['cherry', 'apple', 'pineapple', 'mango']
vegetables = ['Tomato', 'Potato', 'Cabbage',
              'Onion', 'Carrot']      
animal_products = ['milk', 'meat', 'butter',
                   'yoghurt']             
web_techs = ['HTML', 'CSS', 'JS', 'React', 'Redux',
             'Node', 'MongDB']  
countries = ['Finland', 'Estonia', 'Denmark', 'Sweden', 'Norway']

# Print the lists and it length
print('Fruits:', fruits)
print('Number of fruits:', len(fruits))
print('Vegetables:', vegetables)
print('Number of vegetables:', len(vegetables))
print('Animal products:', animal_products)
print('Number of animal products:', len(animal_products))
print('Web technologies:', web_techs)
print('Number of web technologies:', len(web_techs))
print('Number of countries:', len(countries))
````

````python
# Indexing
fruits = ['cherry', 'apple', 'pineapple', 'mango']
first_fruit = fruits[0]  # we are accessing the first item using its index
print(first_fruit)     
second_fruit = fruits[1]
print(second_fruit)     
last_fruit = fruits[3]
print(last_fruit) 
# Last index
last_index = len(fruits) - 1
last_fruit = fruits[last_index]
````

````python
# Negative indexing
fruits = ['cherry', 'apple', 'pineapple', 'mango']
last_fruit = fruits[-1] # we are accessing the last item using its index
second_last = fruits[-2]
print(last_fruit)      
print(second_last)   
````

````python
# index
fruits = ['cherry', 'apple', 'pineapple', 'mango']
print(fruits.index('apple'))   
ages = [22, 19, 24, 25, 26, 24, 25, 24]
print(ages.index(24))
````

````python
# Modifying items
fruits = ['cherry', 'apple', 'pineapple', 'mango']
fruits[0]='coconut'
print(fruits)
````

````python
# Slicing items
fruits = ['cherry', 'apple', 'pineapple', 'mango']
all_fruits = fruits[0:4]  
print("fruits[0:4]",all_fruits)
all_fruits = fruits[0:]  
print("fruits[0:]",all_fruits)
apple_and_pineapple = fruits[1:3] 
print("fruits[1:3]",apple_and_pineapple)
apple_pineapple_mango = fruits[1:]
print("fruits[1:]",apple_pineapple_mango)
````

````python
# Negative slicing
fruits = ['cherry', 'apple', 'pineapple', 'mango']
all_fruits = fruits[-4:] 
print("fruits[-4:]",all_fruits)
apple_and_pineapple = fruits[-3:-1]
print("fruits[-3:-1]",apple_and_pineapple)
apple_pineapple_mango = fruits[-3:]
print("fruits[-3:]",apple_pineapple_mango)
````

````python
# checking items
fruits = ['cherry', 'apple', 'pineapple', 'mango']
does_exist = 'cherry' in fruits
print(does_exist)  
does_exist = 'lime' in fruits
print(does_exist)  
````

````python
# Append
fruits = ['cherry', 'apple', 'pineapple', 'mango']
fruits.append('coconut')
print(fruits)           
fruits.append('lime')
print(fruits)
````

````python
# insert
fruits = ['cherry', 'apple', 'pineapple', 'mango']
fruits.insert(2, 'coconut')  
print(fruits)           
fruits.insert(3, 'lime')
print(fruits)
````

````python
# remove
fruits = ['cherry', 'apple', 'pineapple', 'mango']
fruits.remove('cherry')
print(fruits)  
fruits.remove('mango')
print(fruits)  
````

````python
# delete
fruits = ['cherry', 'apple', 'pineapple', 'mango']
del fruits[0]
print(fruits)       
````

````python
# clear
fruits = ['cherry', 'apple', 'pineapple', 'mango']
fruits.clear()
print(fruits)      
````

````python
# copying a lits

fruits = ['cherry', 'apple', 'pineapple', 'mango']
fruits_copy = fruits.copy()
print(fruits_copy) 
````

````python
# join
positive_numbers = [1, 2, 3, 4, 5]
zero = [0]
negative_numbers = [-5, -4, -3, -2, -1]
integers = negative_numbers + zero + positive_numbers
print(integers)
fruits = ['cherry', 'apple', 'pineapple', 'mango']
vegetables = ['Tomato', 'Potato', 'Cabbage', 'Onion', 'Carrot']
fruits_and_vegetables = fruits + vegetables
print(fruits_and_vegetables)
````

````python
# join with extend
num1 = [0, 1, 2, 3]
num2 = [4, 5, 6]
num1.extend(num2)
print('Numbers:', num1)

negative_numbers = [-5, -4, -3, -2, -1]
positive_numbers = [1, 2, 3, 4, 5]
zero = [0]
negative_numbers.extend(zero)
negative_numbers.extend(positive_numbers)
print('Integers:', negative_numbers)

fruits = ['cherry', 'apple', 'pineapple', 'mango']
vegetables = ['Tomato', 'Potato', 'Cabbage', 'Onion', 'Carrot']
fruits.extend(vegetables)
print('Fruits and vegetables:', fruits)
````

````python
# count
fruits = ['cherry', 'apple', 'pineapple', 'mango']
print(fruits.count('apple'))  
ages = [22, 19, 24, 25, 26, 24, 25, 24]
print(ages.count(24))        
````

````python
# reverse
fruits = ['cherry', 'apple', 'pineapple', 'mango']
fruits.reverse()
print(fruits)
````

````python
# reversed
fruits = ['cherry', 'apple', 'pineapple', 'mango']
reserved_fruits=list(reversed(fruits)) # reversed(fruits) creates an iterator, which is then converted to a list using list()
print(reserved_fruits)
````

````python
#  list comprehensions to reverse a list
fruits = ['cherry', 'apple', 'pineapple', 'mango']
reserved_fruits= [fruit for fruit in fruits [::-1]]
print(reserved_fruits)
````

````python
# list comprehensions to  reverse number
numbers = [1, 2, 3, 4, 5]
number_reversed = [num for num in numbers[::-1]]
print(number_reversed)
````

````python
# list comprehensions to power and reverse number
numbers = [1, 2, 3, 4, 5]
squared_reversed = [num**2 for num in numbers[::-1]]
print(squared_reversed)
````

````python
# list comprehension to apply the power only to even numbers using an if condition inside the comprehension
numbers = [1, 2, 3, 4, 5]
# Square only even numbers, leave odd numbers as they are
squared_reversed = [num**2 if num % 2 == 0 else num for num in numbers[::-1]]
print(squared_reversed)
````

````python
# sort
fruits = ['cherry', 'apple', 'pineapple', 'mango']
fruits.sort()
print(fruits)

fruits.sort(reverse=True)
print(fruits)

ages = [22, 19, 24, 25, 26, 24, 25, 24]
ages.sort()
print(ages)

ages.sort(reverse=True)
print(ages)
````

## 3. Sets

**Definition:**
A set is an unordered collection of unique elements.

**Code Example**
```python
numbers = {1, 2, 3, 3, 4}
```
**Key Features:**
- No duplicates
- Unordered
- Efficient membership testing

**Use cases:**
- Unique identifiers
- Categories
- Membership relations

````python
# empyt set
empty_set = set()
````

````python
# set's length
item_set = {'item1', 'item2', 'item3', 'item4'}
print(item_set)
print(len(item_set))
````

````python
# membership
item_set = {'item1', 'item2', 'item3', 'item4'}
print("Does item_set contain item3? ", 'item3' in item_set)
````

````python
# add
item_set = {'item1', 'item2', 'item3', 'item4'}
item_set.add('item5')
print(item_set)
````

````python
# update - Insert the items from set y into set x
item_set = {'item1', 'item2', 'item3', 'item4'}
item_set.update({'item5','item6','item7'})
print(item_set)
````

````python
# remove 
item_set = {'item1', 'item2', 'item3', 'item4'}
item_set.remove('item1')
print(item_set)
````

````python
# clear
item_set = {'item1', 'item2', 'item3', 'item4'}
print(item_set.clear())
print(item_set)
````

````python
# del
item_set = {'item1', 'item2', 'item3', 'item4'}
del item_set
print(item_set) #NameError: name 'item_set' is not defined
````

````python
# ordering in set
lst = ['item1', 'item2', 'item3', 'item4', 'item1']
st = set(lst)  # the order is random, because sets in general are unordered
print(st)
````

````python
# union
st1 = {'item1', 'item2', 'item3', 'item4'}
st2 = {'item5', 'item6', 'item7', 'item8'}
st3 = st1.union(st2) 
print(st3)
````

````python
# intersection
st4 = {'item1', 'item2', 'item3', 'item4'}
st5 = {'item3', 'item2','item5'}
print(st4.intersection(st5))
````

````python
# subset
st4 = {'item1', 'item2', 'item3', 'item4'}
st5 = {'item3', 'item2','item4'}
print(st5.issubset(st4))
````

````python
# supersubset- Return True if all items set y are present in set x:
st4 = {'item1', 'item2', 'item3', 'item4'}
st5 = {'item3', 'item2','item4'}
print(st5.issuperset(st4))
````

````python
# difference
st4 = {'item1', 'item2', 'item3', 'item4'}
st5 = {'item3', 'item2','item5'}
print(st4.difference(st5))
print(st5.difference(st4))
````

````python
# symmetric difference - a set that contains all items from both set, but not the items that are present in both sets
st4 = {'item1', 'item2', 'item3', 'item4'}
st5 = {'item3', 'item2','item5'}
print(st4.symmetric_difference(st5))
print(st5.symmetric_difference(st4))
````

````python
# disjoint - method returns True if none of the items are present in both sets
st4 = {'item1', 'item2', 'item3', 'item4'}
st5 = {'item3', 'item2','item5'}
print(st4.isdisjoint(st5))
print(st5.isdisjoint(st4))
````

## 4. Dictionaries and Key–Value Modelling

**Definition:**
A dictionary is a mapping from keys to values (**key  &rarr; value**)

**Code Example**
```python
student = {
    "name": "Anna",
    "age": 23,
    "programme": "Computer Science"
```
**Key Features:**
- Keys must be immutable
- Fast lookup

**Use cases:**
- Configuration settings
- Structured records
- JSON-like data modelling

````python
# empty dictionary
empty_dict = {}
````

````python
keys_values = {
    'key1':'value1',
    'key2':'value2', 
    'key3':'value3', 
    'key4':'value4'}
print("len(keys_values)", len(keys_values))
print("keys_values['key1']", keys_values['key1'])
print("keys_values.items()", keys_values.items())
copy_key_values=keys_values.copy()
print(copy_key_values)
````

````python
keys_values = {
    'key1':'value1',
    'key2':'value2', 
    'key3':'value3', 
    'key4':'value4'}
print(keys_values.keys())
print(keys_values.values())
````

````python
# update ()
keys_values = {
    'key1':'value1',
    'key2':'value2', 
    'key3':'value3', 
    'key4':'value4'}
updated_key3= {'key3':'value34'}
keys_values.update(updated_key3)
updated_key1= {'key1':'value3'}
keys_values.update(updated_key1)
print(keys_values)
````

````python
# creating a dictionary from two lists using zip()

# define a list of keys
keys = ['a', 'b', 'c', 'd', 'e']

# define a corresponding list of values
values = [1, 2, 3, 4, 5]

# zip(keys, values) pairs each key with its corresponding value as tuples:
# ('a', 1), ('b', 2), ('c', 3), ...
# the result is an iterator, not a list

# dictionary comprehension:
# iterate over each (key, value) pair produced by zip
# construct a dictionary where each key is mapped to its corresponding value
dict = {k: v for (k, v) in zip(keys, values)}

# print the resulting dictionary
print(dict)
````

````python
# definition by comprehension
squared_dict = {x: x**2 for x in range(1, 6)}
print(squared_dict)
````

````python
# Dictionary Comprehension with Conditional Statements
dict = {x: x**2 for x in range(10) if x**2 % 2 == 0}
print(dict)
````

## 5. Nested Data Structures

**Definition:**
Combinations of lists, dictionaries, tuples, and sets to model complex, real-world data structures.

**Code Example: List of dictionaries**
```python
students = [
    {"name": "Anna", "grade": 1.7},
    {"name": "Ben", "grade": 2.3},
    {"name": "Clara", "grade": 1.3}
]
students[0]["name"]   # "Anna"

course = {
    "title": "Data Science",
    "participants": ["Anna", "Ben", "Clara"]
}

course["participants"][1]  # "Ben"
```
**Use cases:** Student registries, banking systems, research datasets, JSON-like data modeling.

````python
# Flattening a List using comprehension
'''matrix is defined as a 2D list (a list of lists) containing 4 rows of 4 integers each'''
matrix = [
     [9, 3, 8, 3],
     [4, 5, 2, 8],
     [6, 4, 3, 1],
     [1, 0, 4, 5],
 ]
'''generator expression that:
   - iterates through each *row* in the *matrix*
   - for each *row*, iterates through each *item* in that row
   - Returns each individual item, effectively flattening the 2D structure into a 1D sequence'''
print([item for row in matrix for item in row]) # convert the generator to a list using square brackets [] or list()

# Try to run this line , the output will be something like that <generator object <genexpr> at 0x0000015B9C886260>
#print(item for row in matrix for item in row) 
````

````python
# Flattening the list of dictionaries using list comprehension and dict() constructor
list_dictionary = [{'Adam': 50}, {'Sue': 49.5}, {'Alex': 47}] # list of dictionaries
''' Uses a *nested list comprehension* to:
   - Iterate through each dictionary *d* in the list *dict_*
   - For each dictionary, extract all key-value pairs using *d.items()*
   - Create a list of tuples where each tuple contains a key and its corresponding value'''
item_list = [(k, v) for d in list_dictionary for k, v in d.items()] 
# Converts the list of key-value tuples into a single dictionary using the `dict()` constructor.
flat_list = dict(item_list)  
print(flat_list)
````
